```python
def side_branch(x, factor, stage):
    conv_name = "edge_conv_" + str(stage)
    trans_conv_name = "edge_transconv_" + str(stage)
    x = Conv2D(1, (1, 1), activation=None, padding='same', name= conv_name)(x)

    kernel_size = (2*factor, 2*factor)
    x = Conv2DTranspose(1, kernel_size, strides=factor, padding='same',name = trans_conv_name, use_bias=False, activation=None)(x)

    return x
def multi_combine(x):
#     conv_name = "edge_conv_" + str(stage)
    x1 = KL.Conv2D(81, (3,3), activation='relu', padding="same")(x)
    x2 = KL.Conv2D(81, (3,3), activation='relu', padding="same")(x)
    x3 = KL.Conv2D(81, (1,1), activation='relu', padding="same")(x)
    x = KL.concatenate([x1,x2,x3])
    x = KL.Conv2D(81, (2,2), activation='relu', padding="same")(x)
    return x

def build_resblock(input_layer,factor):
    tmp = KL.Conv2D(81, (factor, factor) , padding="same")(input_layer)
    tmp = KL.BatchNormalization(axis=3)(tmp)
    tmp = KL.Activation('relu')(tmp)
    tmp = KL.Conv2D(81, (factor, factor) , padding="same")(tmp)
    tmp = KL.BatchNormalization(axis=3)(tmp)
#     tmp = KL.Activation('relu')(tmp)
    
    tmp = KL.Add()([input_layer,tmp])
#     tmp = KL.BatchNormalization(axis=3)(tmp)
#     tmp = KL.Activation('relu')(tmp)
    
    return tmp
```

```python
def base_model(input_image):
    #resnet
    res_model,C1,C2,C3,C4,C5 = resnet_graph(input_image, "resnet50", stage5=True)
    
    for layer in res_model.layers:
        layer.trainable=False
    
    
    C0 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="fpn_p0")(input_image)
    C1 = BilinearUpSampling2D(target_size=(112, 112))(C1)
    C1 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="fpn_p1")(C1)
    
    P5 = KL.Conv2D(256, (1, 1), name='fpn_c5p5')(C5)
    P4 = KL.Add(name="fpn_p4add")([
        BilinearUpSampling2D(target_size=(14, 14))(P5),
        KL.Conv2D(256, (1, 1), name='fpn_c4p4')(C4)])
    P3 = KL.Add(name="fpn_p3add")([
        BilinearUpSampling2D(target_size=(28, 28))(P4),
        KL.Conv2D(256, (1, 1), name='fpn_c3p3')(C3)])
    P2 = KL.Add(name="fpn_p2add")([
        BilinearUpSampling2D(target_size=(56, 56))(P3),
        KL.Conv2D(256, (1, 1), name='fpn_c2p2')(C2)])
#     P1 = KL.Add(name="fpn_p1add")([
#         BilinearUpSampling2D(target_size=(112, 112))(P2),
#         KL.Conv2D(256, (1, 1), name='fpn_c1p1')(C1)])
    
    # Attach 3x3 conv to all P layers to get the final feature maps.
#     P0_0 = KL.Conv2D(64, (3, 3), activation="relu", padding="SAME", name="fpn_p0_0")(input_image)
#     P0_1 = KL.Conv2D(64, (3, 3), activation="relu", padding="SAME", name="fpn_p0_1")(P0_0)
    
    
    
    P2 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="fpn_p2")(P2)
    P3 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="fpn_p3")(P3)
    P4 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="fpn_p4")(P4)
    P5 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="fpn_p5")(P5)
    
#     P2 = KL.Conv2D(81, (3, 3), activation="softmax", padding="SAME", name="fcn_p2")(P2)
#     P3 = KL.Conv2D(81, (3, 3), activation="softmax", padding="SAME", name="fcn_p3")(P3)
#     P4 = KL.Conv2D(81, (3, 3), activation="softmax", padding="SAME", name="fcn_p4")(P4)
#     P5 = KL.Conv2D(81, (3, 3), activation="softmax", padding="SAME", name="fcn_p5")(P5)
    
#     EC1 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_c1")(C1)
    EC2 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_c2")(C2)
    EC3 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_c3")(C3)
    EC4 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_c4")(C4)
    EC5 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_c5")(C5)
    
#     EP1 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_p1")(P1)
    EP2 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_p2")(P2)
    EP3 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_p3")(P3)
    EP4 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_p4")(P4)
    EP5 = KL.Conv2D(256, (1, 1), padding="SAME", name="edge_p5")(P5)
    
    E0 = KL.Conv2D(81, (1, 1), activation="relu", padding="SAME", name="edge0")(C0)
#     E0_1 = KL.Conv2D(81, (1, 1), activation="relu", padding="SAME", name="edge_p0_1")(P0_1)
    

    E0 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="edge0")(C0)
    E0 = KL.Conv2D(2, (1, 1), activation="relu", padding="SAME", name="edge0_1")(E0)
    
    E1 = KL.Conv2D(256, (3, 3), activation="relu", padding="SAME", name="edge1")(C1)
    E1 = KL.Conv2D(2, (1, 1), activation="relu", padding="SAME", name="edge1_1")(E1)
    
    E2 = KL.Add()([EC2,EP2])
    E2 = KL.Conv2D(256, (3,3), activation="relu", padding="SAME", name="edge2")(E2)
    E2 = KL.Conv2D(2, (1, 1), activation="relu", padding="SAME", name="edge2_1")(E2)
    
    E3 = KL.Add()([EC3,EP3])
    E3 = KL.Conv2D(256, (3,3), activation="relu", padding="SAME", name="edge3")(E3)
    E3= KL.Conv2D(2, (1, 1), activation="relu", padding="SAME", name="edge3_1")(E3)
    
    E4 = KL.Add()([EC4,EP4])
    E4 = KL.Conv2D(256, (3,3), activation="relu", padding="SAME", name="edge4")(E4)
    E4= KL.Conv2D(2, (1, 1), activation="relu", padding="SAME", name="edge4_1")(E4)
    
    E5 = KL.Add()([EC5,EP5])
    E5 = KL.Conv2D(256, (3,3), activation="relu", padding="SAME", name="edge5")(E5)
    E5= KL.Conv2D(2, (1, 1), activation="relu", padding="SAME", name="edge5_1")(E5)
    
    P2 = KL.Conv2D(512, (3, 3), activation="relu", padding="SAME", name="fpn_p2_1")(P2)
    P3 = KL.Conv2D(512, (3, 3), activation="relu", padding="SAME", name="fpn_p3_1")(P3)
    P4 = KL.Conv2D(512, (3, 3), activation="relu", padding="SAME", name="fpn_p4_1")(P4)
    P5 = KL.Conv2D(512, (3, 3), activation="relu", padding="SAME", name="fpn_p5_1")(P5)
        
    
    
    return C0,C1,P2,P3,P4,P5,E0,E1,E2,E3,E4,E5
```

```python
from utils.datagenerator import load_image_gt
import numpy as np
import random
import scipy.misc
from scipy import ndimage
# change RGB to gray scale
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
def mold_image(images, config):
    """Takes RGB images with 0-255 values and subtraces
    the mean pixel and converts it to float. Expects image
    colors in RGB order.
    """
    return images.astype(np.float32) - config.MEAN_PIXEL
```

```python
class MyCallback(keras.callbacks.Callback):
    def __init__(self, model, log = {}):
#         self.model = Model
#         self.mask_loss_name = Model.output_names[0]+"_loss"
#         self.edge_loss_name = Model.output_names[1]+"_loss"
        self.pre_mask_loss = 0
        self.pre_edge_loss = 0
        self.total_mask_gap = 0
        self.total_edge_gap = 0
        self.mask_weight = 1
        self.edge_weight = 0
        print("OK")
    def on_train_begin(self, logs={}):
        print("Starting Training")
        
    
    def on_epoch_begin(self, echo, logs={}):
        trainable_count = int(
            np.sum([K.count_params(p) for p in set(model.trainable_weights)]))
        non_trainable_count = int(
            np.sum([K.count_params(p) for p in set(model.non_trainable_weights)]))      
        
        if self.mask_weight == 1:
            
            print("\x1b[32m\"Train Mask\"\x1b[0m")
            print("\x1b[32m\"Total params: {:,}\"\x1b[0m".format(trainable_count + non_trainable_count))
            print("\x1b[32m\"Trainable params: {:,}\"\x1b[0m".format(trainable_count))
            print("\x1b[32m\"Non-trainable params: {:,}\"\x1b[0m".format(non_trainable_count))
            
        else:
            
            print("\x1b[34m\"Train Edge\"\x1b[0m")
            print("\x1b[34m\"Total params: {:,}\"\x1b[0m".format(trainable_count + non_trainable_count))
            print("\x1b[34m\"Trainable params: {:,}\"\x1b[0m".format(trainable_count))
            print("\x1b[34m\"Non-trainable params: {:,}\"\x1b[0m".format(non_trainable_count))
        model.compile(optimizer=sgd, 
              loss= 'categorical_crossentropy',
              loss_weights=[self.edge_weight,self.edge_weight,self.edge_weight,self.edge_weight*2,self.edge_weight*2,
                           self.mask_weight,self.mask_weight,self.mask_weight,self.mask_weight*2,self.mask_weight*2])
        return
    def on_epoch_end(self, echo, logs={}):
        edge_loss = 0
        mask_loss = 0
        for key in logs:
            if "edge" in key:
                edge_loss += logs.get(key)
            elif "mask" in key:
                mask_loss += logs.get(key)
        self.total_mask_gap += abs(self.pre_mask_loss - mask_loss)
        self.total_edge_gap += abs(self.pre_edge_loss - edge_loss)
        self.pre_mask_loss = float(mask_loss)        
        self.pre_edge_loss = float(edge_loss)
        print("\x1b[31m\"Total Mask Loss GAP : {}\"\x1b[0m".format(self.total_mask_gap))
        print("\x1b[31m\"Total Edge Loss GAP : {}\"\x1b[0m".format(self.total_edge_gap))
        if self.total_mask_gap > self.total_edge_gap:
        # whose loss descend value is smaller or ascend value is bigger, train who
#         if (mask_loss- self.pre_mask_loss) > (edge_loss - self.pre_edge_loss):
            self.mask_weight = 1
            self.edge_weight = 0
            
            #if trained , total gap clear
            self.total_mask_gap = 0
            self.pre_mask_loss = mask_loss
            
            for layer in model.layers:
                if "edge" in layer.name:
                    layer.trainable = False
                if "fcn" in layer.name:
                    layer.trainable = True
        else:
            self.mask_weight = 0
            self.edge_weight = 1
            
            #if trained , total gap clear
            self.total_edge_gap = 0
            self.pre_edge_loss = edge_loss
            
            for layer in model.layers:
                if "edge" in layer.name:
                    layer.trainable = True
                if "fcn" in layer.name:
                    layer.trainable = False
        return
```

```python
def generate_model(input_image):
    #resnet
    res_model,C1,C2,C3,C4,C5 = resnet_graph(input_image, "resnet50", stage5=True)
    
    for layer in res_model.layers:
        layer.trainable=False
    
    P5 = KL.Conv2D(256, (1, 1), name='fpn_c5p5')(C5)
    P4 = KL.Add(name="fpn_p4add")([
        BilinearUpSampling2D(target_size=(14, 14))(P5),
        KL.Conv2D(256, (1, 1), name='fpn_c4p4')(C4)])
    P3 = KL.Add(name="fpn_p3add")([
        BilinearUpSampling2D(target_size=(28, 28))(P4),
        KL.Conv2D(256, (1, 1), name='fpn_c3p3')(C3)])
    P2 = KL.Add(name="fpn_p2add")([
        BilinearUpSampling2D(target_size=(56, 56))(P3),
        KL.Conv2D(256, (1, 1), name='fpn_c2p2')(C2)])
    
    # Attach 3x3 conv to all P layers to get the final feature maps.
    P2 = KL.Conv2D(256, (3, 3), padding="SAME", name="fpn_p2")(P2)
    P3 = KL.Conv2D(256, (3, 3), padding="SAME", name="fpn_p3")(P3)
    P4 = KL.Conv2D(256, (3, 3), padding="SAME", name="fpn_p4")(P4)
    P5 = KL.Conv2D(256, (3, 3), padding="SAME", name="fpn_p5")(P5)
    
#     #for each pyramid generate class heat map using 2*2
    P2 = KL.Conv2D(81, (2, 2), padding="SAME", name="heatmap_p2")(P2)
    P3 = KL.Conv2D(81, (2, 2), padding="SAME", name="heatmap_p3")(P3)
    P4 = KL.Conv2D(81, (1, 1), padding="SAME", name="heatmap_p4")(P4)
    P5 = KL.Conv2D(81, (1, 1), padding="SAME", name="heatmap_p5")(P5)
    
    P5 = BilinearUpSampling2D(target_size=(14, 14))(P5)
    P4 = KL.concatenate([P4,P5])
    P4 = KL.Conv2D(81, (1, 1), padding="SAME", name="Unet_p4")(P4)
    P4 = BilinearUpSampling2D(target_size=(28, 28))(P4)
    P3 = KL.concatenate([P3,P4])
    P3 = KL.Conv2D(81, (1, 1), padding="SAME", name="Unet_p3")(P3)
    P3 = BilinearUpSampling2D(target_size=(56, 56))(P3)
    P2 = KL.concatenate([P2,P3])
#     P2 = KL.Conv2D(81, (2, 2), padding="SAME", name="Unet_p2")(P2) 
    P2 = KL.Conv2D(81, (1, 1), padding="SAME", name="Unet_pconbine")(P2)
#     P2 = Dense(81)(P2)
    
#     P = KL.Conv2D(81, (1, 1), padding="SAME", name="Unet_pconbine")(P2)
#     P = KL.Conv2D(81, (1, 1), padding="SAME", name="Unet_pconbine2")(P)
    PQ = KL.Activation("softmax")(P2)
#     model = KM.Model(inputs=[input_image],outputs = [PQ])
    
    
    return PQ,C1
```
